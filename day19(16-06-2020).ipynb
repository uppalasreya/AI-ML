```python
df=pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/weather_data.csv",parse_dates=['day'])
df
```

```python
df['temperature'].value_counts()
```

```python
data=df.replace({"temperature":-99999,'windspeed':-99999,'event':'0'},np.nan)
data
```

```python
## interpolate()---->only for numerical columns(based on the col value,interpolate fun will come up with a guess value to be filled with NaN)
```

```python
int_data=data.interpolate()
int_data
```

### dropna(): Drops the columns or rows which have missing values

```python
drop_data=data.dropna()    ##it will drop even it have 1missing value
drop_data
```

```python
## if a row have "any" missing values delete the row
drop_data=data.dropna(how='any')
drop_data
```

```python
## drpos which have all missing vaues
drop_data=data.dropna(how='all')
drop_data
```

## Handling missing values using sckit-learn

```python
## simpleImputer
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values =np.nan,strategy='mean')
```

```python
imputer.fit(data.iloc[:,1:3])
```

```python
## transforming the data
data.iloc[:,1:3]=imputer.transform(data.iloc[:,1:3])
data
```
