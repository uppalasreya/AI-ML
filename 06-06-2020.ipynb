```python
import pandas as pd
import numpy as np
s1 = pd.Series([9,11,1, 3, 4, 5, 6, 2, 9]) # Define series 1
s2 = pd.Series([1.1, 3.5, 4.7, 5.8, 2.9, 9.3,1.4,2.9,6.4]) # Define series 2
s3 = pd.Series(['a', 'b', 'c', 'd', 'e','f','g','h','i']) # Define series 3
Data ={'first':s1, 'second':s2, 'third':s3} # Define Data
df = pd.DataFrame(Data) # Create DataFrame
df
```

```python
#Renaming columns
df = df.rename(columns={"second":"Column2","third":"column3"})
df
```

### Changing the row index:
1. set_index()
2. reset_index()

```python
#Column3 data will be set as Row index temporarily
df = df.set_index('column3')
df
```

```python
#undo the changes done for row indexing
df = df.reset_index()
df
```

```python
#Iterable object for Row Indexes
label = list("abcdefghi")
label
```

```python
#Setting Iterable Object values as Row index
df = df.set_index(pd.Index(label))
df
```

### Removing columns:¶

```python
import pandas as pd
import numpy as np
s1 = pd.Series([9,11,1, 3, 4, 5, 6, 2, 9]) # Define series 1
s2 = pd.Series([1.1, 3.5, 4.7, 5.8, 2.9, 9.3,1.4,2.9,6.4]) # Define series 2
s3 = pd.Series(['a', 'b', 'c', 'd', 'e','f','g','h','i']) # Define series 3
Data ={'first':s1, 'second':s2, 'third':s3} # Define Data
df = pd.DataFrame(Data) # Create DataFrame
df
```

```python
#single column removing
df.drop(["first"],axis=1)
```

```python
#Removing multiple columns
df.drop(["first","second"],axis=1)
```

```python
#Removing multiple columns with position
df.drop(df.columns[[0,2]],axis=1)
```

```python
df
```

```python
#Removing single row
df.drop([7],axis=0)
```

```python
#Remove multiple rows
df.drop([7,8])
```
